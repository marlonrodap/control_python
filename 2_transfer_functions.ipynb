```python
import control
import numpy as np
import matplotlib.pyplot as plt
```

We are going to define the transfer function:

$$H\left(s\right)=\frac{3}{4s + 1}$$

```python
num = np.array([3])
den = np.array([4, 1])

H = control.tf(num, den)
print("H(s) =", H)
```

```python
t, y = control.step_response(H)

plt.plot(t,y)
plt.title("Step Response")
plt.grid()
print(len(y))
```

## Now try the example SSM with our 1_ODE example

![SSM1ODE](/Users/marlonrodriguez/Library/CloudStorage/OneDrive-KULeuven/Programming/py/control_python/images/ssm1ode.png "First Order Transfer") 

```python
import scipy.signal as sig
```

```python
# Simulation Parameters
x0 = [0, 0]

start = 0
stop = 30
step = 1
t = np.arange(start, stop, step)

K = 3
T = 4

# State-space Model
A = [[-1/T, 0],
     [0, 0]]
B = [[K/T],
     [0]]
C = [[1, 0]]
D = 0

sys = sig.StateSpace(A, B, C, D)

# Step Response
t, y = sig.step(sys, x0, t)

# Plotting
plt.plot(t,y)
plt.title("Step Response")
plt.xlabel("t [s]")
plt.ylabel("y(t)")
plt.grid()
```

We get the same results if we use the transfer function derived from the State-space Model

```python
H = sys.to_tf()

# Step Response
t, y = sig.step(H, x0, t)

# Plotting
plt.plot(t,y)
plt.title("Step Response")
plt.xlabel("t [s]")
plt.ylabel("y(t)")
plt.grid()
```
